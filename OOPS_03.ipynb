## Overloading Coding Questions

- 1. Overload __add__ for a Fraction class to add two fractions.

```python
class Fraction:
    def __init__(self, numerator, denominator):
        self.numerator = numerator
        self.denominator = denominator

    def __add__(self, other):
        # Formula: (a/b + c/d) = (ad + bc) / (bd)
        new_num = self.numerator * other.denominator + other.numerator * self.denominator
        new_den = self.denominator * other.denominator
        return Fraction(new_num, new_den)

    def __str__(self):
        return f"{self.numerator}/{self.denominator}"


# Example usage:
f1 = Fraction(1, 2)
f2 = Fraction(1, 3)

result = f1 + f2   
print(result)      
```

- 6. Overload __mul__ for a Vector class to support scalar and dot product
multiplication.


```python
class Vector:
    def __init__(self, values):
        self.values = values  # list of numbers

    def __mul__(self, other):
        # Case 1: Scalar multiplication (number * vector)
        if isinstance(other, (int, float)):
            return Vector([x * other for x in self.values])
        
        # Case 2: Dot Product (vector * vector)
        if isinstance(other, Vector):
            if len(self.values) != len(other.values):
                raise ValueError("Vectors must have the same length for dot product.")
            return sum(a * b for a, b in zip(self.values, other.values))
        
        return NotImplemented

    def __repr__(self):
        return f"Vector({self.values})"


# Example Usage
v1 = Vector([1, 2, 3])
v2 = Vector([4, 5, 6])

print(v1 * 3)       # Scalar multiplication ‚Üí Vector([3, 6, 9])
print(v1 * v2)      # Dot product ‚Üí (1*4 + 2*5 + 3*6) = 32
```

- 10. Implement __len__ for a custom Queue class.


```python
class Queue:
    def __init__(self):
        self.items = []

    def enqueue(self, item):
        self.items.append(item)

    def dequeue(self):
        if not self.items:
            return None
        return self.items.pop(0)

    def __len__(self):
        return len(self.items)  # Allows len(queue) to work

    def __str__(self):
        return str(self.items)


# Example Usage
q = Queue()
q.enqueue(10)
q.enqueue(20)
q.enqueue(30)

print("Queue:", q)         # Queue: [10, 20, 30]
print("Length:", len(q))   # Length: 3

q.dequeue()
print("Queue after dequeue:", q)   # Queue after dequeue: [20, 30]
print("Length:", len(q))           # Length: 2
```

- 3. Create a Vehicle class with a start method, override in Car to add oil check and
call parent.


```python
# Parent class
class Vehicle:
    def __init__(self, name):
        self.name = name

    def start(self):
        print(f"{self.name} is starting.")


# Subclass
class Car(Vehicle):
    def __init__(self, name, oil_level):
        super().__init__(name)   # Call parent __init__
        self.oil_level = oil_level

    def start(self):
        # Perform additional check before starting
        if self.oil_level < 20:
            print(f"Warning: Low oil level ({self.oil_level}%). Cannot start {self.name}.")
        else:
            print(f"Oil level is sufficient ({self.oil_level}%).")
            super().start()  # Call parent start method


# Example Usage
v = Vehicle("Generic Vehicle")
v.start()   # Output: Generic Vehicle is starting.

c1 = Car("Honda Civic", 50)
c1.start()
# Output:
# Oil level is sufficient (50%).
# Honda Civic is starting.

c2 = Car("Toyota Corolla", 10)
c2.start()
# Output:
# Warning: Low oil level (10%). Cannot start Toyota Corolla.
```

- 5. Create an Account base class with withdraw , override in SavingsAccount with a
limit.


```python
# Base class
class Account:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance

    def withdraw(self, amount):
        if amount > self.balance:
            print(f"Insufficient balance. Current balance: {self.balance}")
        else:
            self.balance -= amount
            print(f"Withdrawn {amount}. New balance: {self.balance}")


# Subclass with withdrawal limit
class SavingsAccount(Account):
    def __init__(self, owner, balance, min_balance):
        super().__init__(owner, balance)
        self.min_balance = min_balance

    def withdraw(self, amount):
        if self.balance - amount < self.min_balance:
            print(f"Cannot withdraw {amount}. Minimum balance of {self.min_balance} must be maintained.")
        else:
            super().withdraw(amount)  # Call parent method


# Example Usage
a = Account("Alice", 1000)
a.withdraw(200)  # Withdrawn 200. New balance: 800

s = SavingsAccount("Bob", 1000, 500)
s.withdraw(400)  # Cannot withdraw 400. Minimum balance of 500 must be maintained.
s.withdraw(300)  # Withdrawn 300. New balance: 700
```

- 1. Create a Computer class that composes CPU , RAM , and Storage classes.

```python
# Component classes
class CPU:
    def __init__(self, model, cores):
        self.model = model
        self.cores = cores

    def __str__(self):
        return f"CPU(model: {self.model}, cores: {self.cores})"


class RAM:
    def __init__(self, size_gb):
        self.size_gb = size_gb

    def __str__(self):
        return f"RAM({self.size_gb}GB)"


class Storage:
    def __init__(self, size_gb, type_):
        self.size_gb = size_gb
        self.type_ = type_

    def __str__(self):
        return f"Storage({self.size_gb}GB {self.type_})"


# Computer class using composition
class Computer:
    def __init__(self, cpu, ram, storage):
        self.cpu = cpu
        self.ram = ram
        self.storage = storage

    def show_specs(self):
        print(self.cpu)
        print(self.ram)
        print(self.storage)


# Example Usage
cpu = CPU("Intel i7", 8)
ram = RAM(16)
storage = Storage(512, "SSD")

my_pc = Computer(cpu, ram, storage)
my_pc.show_specs()
```

- 2. Create a Car class that composes Engine and Transmission .


```python
# Component classes
class Engine:
    def __init__(self, model, horsepower):
        self.model = model
        self.horsepower = horsepower

    def start(self):
        print(f"Engine {self.model} with {self.horsepower} HP is starting.")

    def __str__(self):
        return f"Engine(model: {self.model}, HP: {self.horsepower})"


class Transmission:
    def __init__(self, type_):
        self.type_ = type_

    def shift(self):
        print(f"Transmission ({self.type_}) is shifting gears.")

    def __str__(self):
        return f"Transmission(type: {self.type_})"


# Car class using composition
class Car:
    def __init__(self, make, engine, transmission):
        self.make = make
        self.engine = engine
        self.transmission = transmission

    def start_car(self):
        print(f"Starting {self.make} car...")
        self.engine.start()
        self.transmission.shift()

    def show_specs(self):
        print(f"Car make: {self.make}")
        print(self.engine)
        print(self.transmission)


# Example Usage
engine = Engine("V8", 450)
transmission = Transmission("Automatic")
car = Car("Ford Mustang", engine, transmission)

car.show_specs()
car.start_car()
```

- 3. Create a Playlist class that composes a list of Song objects.


```python
# Component class
class Song:
    def __init__(self, title, artist, duration):
        self.title = title
        self.artist = artist
        self.duration = duration  # in seconds

    def __str__(self):
        minutes = self.duration // 60
        seconds = self.duration % 60
        return f"{self.title} by {self.artist} [{minutes}:{seconds:02d}]"


# Playlist class using composition
class Playlist:
    def __init__(self, name):
        self.name = name
        self.songs = []  # List to hold Song objects

    def add_song(self, song):
        self.songs.append(song)

    def remove_song(self, song):
        if song in self.songs:
            self.songs.remove(song)

    def show_playlist(self):
        print(f"Playlist: {self.name}")
        for idx, song in enumerate(self.songs, start=1):
            print(f"{idx}. {song}")


# Example Usage
song1 = Song("Shape of You", "Ed Sheeran", 240)
song2 = Song("Blinding Lights", "The Weeknd", 200)
song3 = Song("Levitating", "Dua Lipa", 220)

my_playlist = Playlist("My Favorites")
my_playlist.add_song(song1)
my_playlist.add_song(song2)
my_playlist.add_song(song3)

my_playlist.show_playlist()
```

- 8. Create an Order class that composes a Customer and a list of Item objects

```python
# Component classes
class Customer:
    def __init__(self, name, email):
        self.name = name
        self.email = email

    def __str__(self):
        return f"Customer: {self.name}, Email: {self.email}"


class Item:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def __str__(self):
        return f"{self.name} (${self.price})"


# Order class using composition
class Order:
    def __init__(self, customer):
        self.customer = customer
        self.items = []  # List to hold Item objects

    def add_item(self, item):
        self.items.append(item)

    def remove_item(self, item):
        if item in self.items:
            self.items.remove(item)

    def total_amount(self):
        return sum(item.price for item in self.items)

    def show_order(self):
        print(self.customer)
        print("Order Items:")
        for item in self.items:
            print(f"- {item}")
        print(f"Total Amount: ${self.total_amount()}")


# Example Usage
customer = Customer("Husna", "husna@example.com")
item1 = Item("Laptop", 1000)
item2 = Item("Mouse", 50)
item3 = Item("Keyboard", 80)

order = Order(customer)
order.add_item(item1)
order.add_item(item2)
order.add_item(item3)

order.show_order()
```

- 9. Show aggregation with a Team class containing Player objects that can exist
independently.


```python
# Component class
class Player:
    def __init__(self, name, position):
        self.name = name
        self.position = position

    def __str__(self):
        return f"{self.name} ({self.position})"


# Aggregating class
class Team:
    def __init__(self, name):
        self.name = name
        self.players = []  # List to hold Player objects

    def add_player(self, player):
        self.players.append(player)

    def remove_player(self, player):
        if player in self.players:
            self.players.remove(player)

    def show_team(self):
        print(f"Team: {self.name}")
        for player in self.players:
            print(f"- {player}")


# Example Usage
player1 = Player("Alice", "Forward")
player2 = Player("Bob", "Goalkeeper")
player3 = Player("Charlie", "Midfielder")

# Players exist independently
print(player1)  # Alice (Forward)
print(player2)  # Bob (Goalkeeper)

# Aggregation: create a team and add players
team = Team("Eagles")
team.add_player(player1)
team.add_player(player2)
team.add_player(player3)

team.show_team()
```

- 10. Create a Menu class that composes MenuItem objects recursively.

```python
# Component class
class MenuItem:
    def __init__(self, name, price=0, submenu=None):
        self.name = name
        self.price = price
        self.submenu = submenu  # Can be another Menu object

    def __str__(self):
        return f"{self.name} (${self.price})" if self.price > 0 else self.name


# Menu class using recursive composition
class Menu:
    def __init__(self, title):
        self.title = title
        self.items = []

    def add_item(self, menu_item):
        self.items.append(menu_item)

    def show_menu(self, indent=0):
        prefix = " " * indent
        print(f"{prefix}{self.title}")
        for item in self.items:
            print(f"{prefix}- {item}")
            if item.submenu:
                item.submenu.show_menu(indent + 4)  # Recursive call for submenu


# Example Usage
desserts_menu = Menu("Desserts")
desserts_menu.add_item(MenuItem("Ice Cream", 5))
desserts_menu.add_item(MenuItem("Cake", 6))

drinks_menu = Menu("Drinks")
drinks_menu.add_item(MenuItem("Coffee", 3))
drinks_menu.add_item(MenuItem("Tea", 2))

main_menu = Menu("Main Menu")
main_menu.add_item(MenuItem("Burger", 8))
main_menu.add_item(MenuItem("Pizza", 10))
main_menu.add_item(MenuItem("Desserts", submenu=desserts_menu))
main_menu.add_item(MenuItem("Drinks", submenu=drinks_menu))

main_menu.show_menu()
```
